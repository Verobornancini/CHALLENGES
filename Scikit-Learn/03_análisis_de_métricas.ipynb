# Scikit Learn: step 3
## Challenge: Credit card fraud detection (Keagle)

```python
# Importo librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

# obtengo el dataset descargado de la plataforma keagle: https://www.kaggle.com/mlg-ulb/creditcardfraud
df  = pd.read_csv("creditcard.csv")[:80_000] # como es demasiado grande, solo leeremos los primeros 80 mil casos
print (df.shape)
df.head(3)
```

```python
# Armo mis dos conjuntos de datos, X: variables predictoras, y: variable target que quiero predecir
X = df.drop(columns=['Time', 'Amount', 'Class']).values
y = df['Class'].values
print (f"Shape of X = {X.shape}")
print (f"Shape of y = {y.shape}")
print (f"Fraud Cases = {y.sum()}")
```

* Es notorio que el número de casos de fraudes es mucho menor que la cantidad de casos registrados,
* Nuestro dataset está desbalanceado,

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000) # necesito setearlo porque no alcanzan las iteraciones que realiza por defecto, por el desbalance de datos

model_fit = model.fit(X, y).predict(X)

print(f" Fraud cases predicted = {model_fit.sum()}")
```

* Tengo menos casos de fraudes detectados que los que observo en la base,
* Para mejorar el modelo puedo ajustarlo, asignandole diferente peso a las clases.
* clase 0: 1, clase 1: 2

```python
model = LogisticRegression(class_weight = {0: 1, 1: 2},# asigno diferentes pesos a cada clase
                           max_iter = 1000) # necesito setearlo porque no alcanzan las iteraciones que realiza por defecto, por el desbalance de datos

model_fit = model.fit(X, y).predict(X)

print(f" Fraud cases predicted = {model_fit.sum()}")
```

* Observamos que mejora la cantidad de casos de fraudes predichos,
* Buscaremos con GridSerch los mejores valores para los pesos de las clases. 

```python
# importo la grilla
from sklearn.model_selection import GridSearchCV
# importo las metricas de scoring
from sklearn.metrics import make_scorer
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

grid = GridSearchCV(
    estimator = LogisticRegression(max_iter = 1000),
    #param_grid = {'class_weight': [{0: 1, 1: v} for v in range(1, 4)]}, # probaremos varios valores de pesos para la clase 1
    param_grid = {'class_weight': [{0: 1, 1: v} for v in np.linspace(1,20, 30)]}, # con linspace puedo lograr una mejor resolución
    scoring = {'precision': make_scorer(precision_score), 
               'recall': make_scorer(recall_score)},
    refit = 'precision', # Si quiero que Gridearch elija el mejor modelo basado en uno de los scores, le indico cual 
    return_train_score = True,
    cv = 10, # cantidad de grupos para cross validation, los scrores obtenids para cada grupo se usan para asignar el rank_test_score
    n_jobs = -1
)

# Entreno el modelo
grid.fit(X, y)

# Analizo todos los resultados del cross validation
pd.DataFrame(grid.cv_results_) # convierto en dataframe al diccionario de resultados para verlo mejor
```

```python
# Gráfico para el set de Test o Validación
plt.figure(figsize=(12, 4))
df_results = pd.DataFrame(grid.cv_results_)
for score in ['mean_test_recall', 'mean_test_precision']:
    plt.plot([_[1] for _ in df_results['param_class_weight']], 
             df_results[score], 
             label=score)
plt.legend();
```

* eje X = Class Weight
* Eje Y = Scores recall and precission

* Nota: El modelo con mejor precisión está en el extremo izquierdo del gráfico, y el con mejor recall en el extremo opuesto.
* Obsrvamos que será util usar una métrica que combien a las dos anteriores y nos hable de un balance: f1

```python
# Gráfico para el set de Entrenamiento
plt.figure(figsize=(12, 4))
df_results = pd.DataFrame(grid.cv_results_)
for score in ['mean_train_recall', 'mean_train_precision']:
    plt.plot([_[1] for _ in df_results['param_class_weight']], 
             df_results[score], 
             label=score)
plt.legend();
```

* Crearemos una nueva función que nos devuelva el valor mínimo entre recall y precisión para obtener el modelo mejor balanceado.

```python
# función definida usando las funciones de precisión y recall que ya trae Scikit-Learn
def min_recall_precision(y_true, y_pred):
    recall = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    return min(recall, precision)

# agrego la función anteriormente creada al GridSearch
grid = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000),
    param_grid={'class_weight': [{0: 1, 1: v} for v in np.linspace(1, 20, 30)]},
    scoring={'precision': make_scorer(precision_score), 
             'recall': make_scorer(recall_score),
             'min_both': make_scorer(min_recall_precision)}, # función que definí arriba
    refit='min_both', # diremos que ajuste el mejor modelo sobre este nuevo scrore que construimos
    return_train_score=True,
    cv=10,
    n_jobs=-1
)
grid.fit(X, y);
```

```python
plt.figure(figsize=(12, 4))
df_results = pd.DataFrame(grid.cv_results_)
for score in ['mean_train_recall', 'mean_train_precision', 'mean_test_min_both']:
    plt.plot([_[1] for _ in df_results['param_class_weight']], 
             df_results[score], 
             label=score)
plt.legend();
```

# A partir de acá, seguiremos una nueva aproximación
* Podemos seguir la aproximación de interpretar a los fraudes como outlayers

```python
from collections import Counter
from sklearn.ensemble import IsolationForest

mod = IsolationForest().fit(X) # IsolationForest necesitan que le pasen un dataframe completo: X
np.where(mod.predict(X) == -1, 1, 0) # transformamos los datos de resultado -1 para que sean 1 (fraudes) y 0 (no fraudes)

# esa transformación es porque los scores de precisión y recall necesitan valores de 1 y 0
# en base a este dato, creamos nuestos scores propios para outlayers

def outlier_precision(mod, X, y):
    preds = mod.predict(X)
    return precision_score(y, np.where(preds == -1, 1, 0))

def outlier_recall(mod, X, y):
    preds = mod.predict(X)
    return recall_score(y, np.where(preds == -1, 1, 0))
```

```python
grid = GridSearchCV(
    estimator=IsolationForest(),
    param_grid={'contamination': np.linspace(0.001, 0.02, 10)}, # parametro para optimizar
    scoring={'precision': outlier_precision, 
             'recall': outlier_recall},
    refit='precision',
    cv=5,
    n_jobs=-1
)
grid.fit(X, y);

plt.figure(figsize=(12, 4))
df_results = pd.DataFrame(grid.cv_results_)
for score in ['mean_test_recall', 'mean_test_precision']:
    plt.plot(df_results['param_contamination'], 
             df_results[score], 
             label=score)
plt.legend();
```
